# Discentes por programa

```python
%load_ext sql
```

```python
%%sql postgresql://capesdb_user:c4pes@capesdb:5432/capesdb

    SELECT 
        NM_PROGRAMA_IES AS PROGRAMA
        ,CD_PROGRAMA_IES
        ,COUNT(ID_PESSOA) NR_DISCENTES
        ,AN_BASE
FROM FATO_DISCENTE
WHERE 
        NM_SITUACAO_DISCENTE = 'MATRICULADO'
        AND SG_ENTIDADE_ENSINO = 'UFPA'
        AND AN_BASE <= 2016
        AND cd_programa_ies = '15001016004P8'
GROUP BY PROGRAMA, CD_PROGRAMA_IES	,AN_BASE
ORDER BY CD_PROGRAMA_IES ASC, AN_BASE ASC --MANDATORY
```

```python
%%sql
    SELECT DISTINCT(ID_PESSOA)
        ,AN_BASE
        ,SG_ENTIDADE_ENSINO
        ,NM_PROGRAMA_IES
        ,ID_PESSOA
        ,NM_DISCENTE
    FROM FATO_DISCENTE
    WHERE cd_programa_ies = '15001016004P8' 
    AND AN_BASE = 2016
    ORDER BY NM_DISCENTE ASC
```
