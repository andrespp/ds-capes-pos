# UFPA

Cursos de Pós graduação da Universidade Federal do Pará

## 2016

### Ranking

```python
%load_ext sql
```

```python
%%sql postgresql://capesdb_user:c4pes@capesdb:5432/capesdb
            
    SELECT * 
    FROM UFPA_EVALUATION 
    WHERE AN_BASE = 2016
    ORDER BY 
        CD_CONCEITO_PROGRAMA DESC
        ,NR_DOUTORES_TITULADOS DESC
        ,NR_MESTRES_TITULADOS DESC
```

### Docentes PPGEE

```python
%%sql postgresql://capesdb_user:c4pes@capesdb:5432/capesdb
    SELECT COUNT(DISTINCT ID_PESSOA) AS NUMERO_DE_DOCENTES
    FROM FATO_DOCENTE 
    WHERE 
        SG_ENTIDADE_ENSINO = 'UFPA' 
        AND NM_PROGRAMA_IES = 'ENGENHARIA ELÉTRICA'
        AND AN_BASE = 2016
```

```python
%%sql postgresql://capesdb_user:c4pes@capesdb:5432/capesdb
   SELECT DISTINCT 
        ID_PESSOA
        ,NM_DOCENTE
        ,AN_BASE
        ,CD_CAT_BOLSA_PRODUTIVIDADE
    FROM FATO_DOCENTE 
    WHERE 
        SG_ENTIDADE_ENSINO = 'UFPA' 
        AND NM_PROGRAMA_IES = 'ENGENHARIA ELÉTRICA'
        AND AN_BASE = 2016
    ORDER BY NM_DOCENTE ASC
```
